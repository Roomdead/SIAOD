### Задание №1
Массив A состоит из целых положительных чисел длин отрезков. Составьте из трех отрезков такой треугольник, чтобы его периметр был максимально возможным. Если невозможно составить треугольник с положительной площадью функция возвращает 0.

```python
num = int(input())
A = []
for i in range(num):
    A.append(int(input()))
A.sort()
fl = True
i = len(A) - 1
while fl and i > 1:
    if A[i] ** 2 <= A[i - 1] ** 2 + A[i - 2] ** 2:
        fl = False
    i -= 1
if fl:
    print(0)
else:
    print(A[i + 1], "+", A[i], "+", A[i - 1], "=", A[i + 1] + A[i] + A[i - 1])
```

### Задание №2
Дан массив неотрицательных целых чисел nums. Расположите их в таком порядке, чтобы вместе они образовали максимально возможное число. Замечание: Результат может быть очень большим числом, поэтому представьте его как string, а не integer.

```python
import random
import sys


def coounter(num):
    s = 0
    while num > 0:
        s += 1
        num //= 10
    return s


def LCM(num1, num2):
    i = 1
    not_calculated = True
    if num1 > num2:
        num1, num2 = num2, num1
    while not_calculated:
        if num1 * i % num2 == 0:
            not_calculated = False
        else:
            i += 1
    return num1 * i


def max_num_sort(arr):
    subarr = []
    lower = []
    eq = []
    hi = []
    #print("inp arr", arr)
    if arr != subarr:
        r_index = random.randint(0, len(arr) - 1)
        #print('r_el from arr', arr[r_index], end="\n\n")
        for el in arr:
            fl = True
            count = 0
            arr_count = coounter(arr[r_index])
            el_count = coounter(el)
            while fl:
                hier_d_el = (el // 10 ** (el_count - 1 - count % el_count)) % 10
                hier_d_r_ind = (arr[r_index] // 10 ** (arr_count - 1 - count % arr_count)) % 10
                #print(hier_d_el,  hier_d_r_ind)
                fl = False
                if hier_d_el == hier_d_r_ind and count < LCM(arr_count, el_count):
                    count += 1
                    fl = True
                elif hier_d_el < hier_d_r_ind:
                    lower.append(el)
                elif hier_d_el > hier_d_r_ind:
                    hi.append(el)
                else:
                    if arr_count > el_count:
                        hi.append(el)
                    elif arr_count < el_count:
                        lower.append(el)
                    else:
                        eq.append(el)
            #print("lower =", lower)
            #print("eq =", eq)
            #print("hi =", hi, end="\n\n")
        subarr = max_num_sort(hi) + eq + max_num_sort(lower)
    return subarr


lgth = int(input('Type the length of input array: '))
rand_array = []
for i in range(lgth):
    newel = random.randint(0, sys.maxsize)
    rand_array.append(newel)
print("Input array is: ", end='')
print(rand_array, end='\n\n')
output_arr = max_num_sort(rand_array)
print("Sorted array is: ", end='')
print(output_arr, end='\n\n')
print("Final number is: ", end="")
for el in output_arr:
    print(el, end="")
input()

```

### Задание №3
Дана матрица mat размером m * n, значения целочисленные. Напишите функцию, сортирующую каждую диагональ матрицы по возрастанию и возвращающую получившуюся матрицу.

```python
import random
import copy


def print_m(matrix):
    m = len(matrix)
    for i in range(m):
        print(matrix[i])


def quickSort(arr):
    if len(arr) <= 1:
        return arr
    startElement = arr[0]
    lowArr = list(filter(lambda x: x < startElement, arr))
    midArr = [i for i in arr if i == startElement]
    highArr = list(filter(lambda x: x > startElement, arr))
    return quickSort(lowArr) + midArr + quickSort(highArr)


def gen_m():
    m = int(input("Enter m: "))
    n = int(input("Enter n: "))
    max_num = int(input("Enter max number value: "))
    matrix = []
    for i in range(m):
        line = []
        for j in range(n):
            line.append(random.randint(0, max_num))
        matrix.append(line)
    print("generated matrix is:")
    print_m(matrix)
    return matrix


def diag_sort(matrix):
    m = len(matrix)
    n = len(matrix[0])
    srtd_matrix = copy.copy(matrix)
    sorted = []
    for k in range(m):
        to_sort = []
        i = k
        filled = False
        while i < m and not filled:
            if i - k < n:
                to_sort.append(matrix[i][i - k])
                i += 1
            else:
                filled = True
        sorted.clear()
        sorted = quickSort(to_sort)
        i = k
        filled = False
        while i < m and not filled:
            if i - k < n:
                j = i - k
                srtd_matrix[i][j] = sorted[j]
                i += 1
            else:
                filled = True
    for k in range(1, n):
        to_sort = []
        i = k
        filled = False
        while i < n and not filled:
            if i - k < m:
                to_sort.append(matrix[i - k][i])
                i += 1
            else:
                filled = True
        sorted.clear()
        sorted = quickSort(to_sort)
        i = k
        filled = False
        while i < n and not filled:
            if i - k < m:
                srtd_matrix[i - k][i] = sorted[i - k]
                i += 1
            else:
                filled = True
    return srtd_matrix


m_1 = gen_m()
print("sorted matrix is:")
print_m(diag_sort(m_1))
```

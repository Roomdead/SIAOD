### Задание №2
Написать генератор случайных матриц(многомерных), который принимает
опциональные параметры <b>m</b>, <b>n</b>, <b>min_limit</b>, <b>max_limit</b>, где <b>m</b> и <b>n</b> указывают размер
матрицы, а <b>min_lim</b> и <b>max_lim</b> - минимальное и максимальное значение для
генерируемого числа.

```python
import random
import time
```

```python
user_m = int(input())
user_n = int(input())
user_min_limit = int(input())
user_max_limit = int(input())
matrix = []

for i in range(user_m):
    row = []
    matrix.append(row)
    for j in range(user_n):
        matrix[i].append(random.randint(user_min_limit, user_max_limit))


def print_m(arr):
    for line in arr:
        print(list(line))


print_m(matrix)
```

### Задание №3
Реализовать методы сортировки строк числовой матрицы в соответствии с
заданием. Оценить время работы каждого алгоритма сортировки и сравнить его со
временем стандартной функции сортировки. Испытания проводить на сгенерированных
матрицах.

```python
import copy
```

```python
# Сортировка выбором.
# Берётся срез массива, в котором минимальный элемент переносят в самый левый угол,
# после чего срез уменьшается и цикл повторяется.
times = []
start_time = time.time()

arr_1 = copy.copy(matrix)
lenght = len(arr_1[0])
for row in arr_1:
    for i in range(lenght - 1):
        min_i = i
        fl = False
        for j in range(i+1, lenght):
            if row[min_i] > row[j]:
                min_i = j
                fl = True
        if fl:
            buf = row[i]
            row[i] = row[min_i]
            row[min_i] = buf
print_m(arr_1)
select_time = round((time.time() - start_time)*1000) 
print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
times.append(["select", select_time])
```

```python
# Сортировка вставкой.
start_time = time.time()
arr_2 = copy.copy(matrix)

for row in arr_2:
    lenght = len(row)
    for i in range(1, lenght):
        fl = True
        j = i
        while fl:
            if row[j] < row[j-1]:
                 row[j], row[j-1] = row[j-1], row[j]
            else:
                fl = False
            j -= 1
            if j == 0:
                fl = False
print_m(arr_2)
inject_time = round((time.time() - start_time)*1000)
print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
times.append(["inject", inject_time])
```

```python
# Сортировка обменом. || Сортировка пузырьком.
start_time = time.time()
arr_3 = copy.copy(matrix)
lenght = len(arr_3[0])
for row in arr_3:
    for i in range(lenght - 1):
        for j in range(0, lenght - 1 - i):
            if row[j] > row[j + 1]:
                row[j], row[j + 1] = row[j + 1], row[j]
print_m(arr_3)
bubble_time = round((time.time() - start_time)*1000)
print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
times.append(["bubble", bubble_time])
```

```python
# Сортировка Шелла.
start_time = time.time()
arr_4 = copy.copy(matrix)

for row in arr_4:
    gap = len(row) // 2
    while gap > 0:
        for i in range(gap, len(row)):
            curr_el = row[i]
            curr_pos = i
            while curr_pos >= gap and row[curr_pos - gap] > curr_el:
                row[curr_pos] = row[curr_pos - gap]
                curr_pos -= gap
            row[curr_pos] = curr_el
        gap //= 2
print_m(arr_4)
shell_time = round((time.time() - start_time)*1000)
print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
times.append(["shell", shell_time])
```

```python
# Быстрая сортировка.
def quickSort(arr):
    if len(arr) > 1:
        x = arr[random.randint(0, len(arr) - 1)]
        low, eq, hi = [], [x], []
        for item in arr:
            if item == x:
                eq.append(item)
            elif item > x:
                hi.append(item)
            else:
                low.append(item)
        return quickSort(low) + eq + quickSort(hi)
    else:
        return arr
    
start_time = time.time()
arr_5 = copy.copy(matrix)
for row in arr_5:
    quickSort(arr_5)
print_m(arr_5)
quick_time = round((time.time() - start_time)*1000)
print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
times.append(["quick", quick_time])    
```

```python
# Турнирная сортировка.
def tournament_sort(alist):
    build_max_heap(alist)
    for i in range(len(alist) - 1, 0, -1):
        alist[0], alist[i] = alist[i], alist[0]
        max_heapify(alist, index=0, size=i)


def parent(i):
    return i // 2


def left(i):
    return 2 * i


def right(i):
    return 2 * i + 1


def build_max_heap(alist):
    length = len(alist)
    start = parent(length)
    while start >= 0:
        max_heapify(alist, index=start, size=length)
        start = start - 1


def max_heapify(alist, index, size):
    l = left(index)
    r = right(index)
    if (l < size and alist[l] > alist[index]):
        largest = l
    else:
        largest = index
    if (r < size and alist[r] > alist[largest]):
        largest = r
    if (largest != index):
        alist[largest], alist[index] = alist[index], alist[largest]
        max_heapify(alist, largest, size)
start_time = time.time()
arr_6 = copy.copy(matrix)
for row in arr_6:
    tournament_sort(row)
print_m(arr_6)
tournament_time = round((time.time() - start_time)*1000)
print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
times.append(["tournament", tournament_time]) 
```
